```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/partB.csv')
```

```python
# df
```

```python
plt.scatter(df['y'],df['V'])
plt.xlabel('d (cm)')
plt.ylabel('Potential (V)')
plt.show()
```

```python
v = [0.5,1,1.5,1.75,2]
```

```python
circle1 = plt.Circle((0, 0), 2, color='grey',label='inner cylinder')
plt.gca().add_patch(circle1)

for i in range(len(v)):
    plt.plot(df[f'x{i+1}'],df[f'y{i+1}'],label=f'{v[i]} V')
    plt.scatter(df[f'x{i+1}'],df[f'y{i+1}'],s=4)
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.legend()
plt.show()
```
