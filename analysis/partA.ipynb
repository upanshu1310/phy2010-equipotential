```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
df = pd.read_csv('../data/partA.csv')
```

```python
df
```

```python
plt.scatter(df['y']+7,df['V'],s=16,label='data',zorder=2)
m,c = np.polyfit(df['y']+7,df['V'],1)
plt.plot(df['y']+7,m*(df['y']+7)+c,label='polyfit',color='red',zorder=1)
plt.xlabel('d (cm)')
plt.ylabel('Potential (V)')
plt.legend()
plt.show()
print(m,c)
```

```python
v = [1,1.5,2,2.5,1.5]
e = 0.2
```

```python
rectangle1 = patches.Rectangle((-11, -7), 22, 1, color="grey",label='electrode')
rectangle2 = patches.Rectangle((-11, 7), 22, 1, color="grey")
for i in range(1,5):
    # plt.errorbar(df[f'x{i}'],df[f'y{i}'], xerr=e, yerr=e,fmt='None',color='black',zorder=1)
    
    plt.scatter(df[f'x{i}'],df[f'y{i}'],s=16)
    plt.plot(df[f'x{i}'],df[f'y{i}'],label=f'{v[i-1]} V')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.gca().add_patch(rectangle1)
plt.gca().add_patch(rectangle2)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=5)

# Adjust the layout to make room for the legend
plt.tight_layout()
plt.show()
```
