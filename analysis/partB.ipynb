```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/partB.csv')
```

```python
df
```

```python
plt.scatter(np.log(abs(df['y'])/2),df['V']-2.65,s=16,label='data',zorder=2)
m,c = np.polyfit(np.log(abs(df['y'])/2),df['V']-2.65,1)
plt.plot(np.log(abs(df['y'])/2),m*(np.log(abs(df['y'])/2))+c,color='red',label='polyfit',zorder=1)
plt.xlabel('log(r/r1)')
plt.ylabel('Potential (V)')
plt.legend()
plt.show()
print(m,c)
print()
```

```python
v = [0.5,1,1.5,1.75,2]
```

```python
circle1 = plt.Circle((0, 0), 2, color='grey',label='electrode')
circle2 = plt.Circle((0, 0), 10, color='grey',fill=False,linewidth=5)
plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)

for i in range(len(v)):
    plt.plot(df[f'x{i+1}'],df[f'y{i+1}'],label=f'{v[i]} V')
    plt.scatter(df[f'x{i+1}'],df[f'y{i+1}'],s=4)
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=6)

# Adjust the layout to make room for the legend
plt.tight_layout()
plt.show()
```
